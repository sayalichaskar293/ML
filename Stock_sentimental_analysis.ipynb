```python
train=df[df["Date"]< "20150101"]
test=df[df["Date"]> "20141231"]
```

```python
data=train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)

#Renaming column names
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns =new_index
data.head(5)
```

```python
#converting to lowercase
for index in new_index:
  data[index]=data[index].str.lower()

data.head(2)
```

```python
" ".join(str(x) for x in data.iloc[1, 0:25])
```

```python
headlines=[]
for row in range(0, len(data.index)):
  headlines.append(" ".join(str(x) for x in data.iloc[row, 0:25]))
```

```python
import matplotlib.pyplot as plt
```

```python
text_for_wc=" ".join(headlines)
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate(text_for_wc)
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#BOW
countVector=CountVectorizer(ngram_range=(2,2))
traindataset=countVector.fit_transform(headlines)
```

```python
# Drop rows with NaN values in specific columns
# train.dropna(subset=["Label"], inplace=True)
```

```python
#randomForest
rfClassifier=RandomForestClassifier(n_estimators=200, criterion="entropy")
# rfClassifier.fit(traindataset, train["Label"])
```

```python
rfClassifier.fit(traindataset, train["Label"])
```

```python
test_headlines=[]
for row in range(0, len(test.index)):
  test_headlines.append(" ".join(str(x) for x in test.iloc[row, 2:27]))
testdataset= countVector.transform(test_headlines)
predictions=rfClassifier.predict(testdataset)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix=confusion_matrix(test["Label"], predictions)
print(matrix)
score=accuracy_score(test["Label"], predictions)
print(score)
report=classification_report(test["Label"], predictions)
print(report)
```
