```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import re
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0, len(news)):
  cleaned=re.sub("[^a-zA-Z]", " ", news["title"][i])
  cleaned=cleaned.lower()
  cleaned=cleaned.split()

  cleaned=[ps.stem(word) for word in cleaned if not word in stopwords.words("english")]
  cleaned=" ".join(cleaned)
  corpus.append(cleaned)

```

```python
corpus
```

```python
X=news.drop("label", axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, ngram_range=(1,3))
X=cv.fit_transform(corpus).toarray()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

```

**NAIVE BAYES ALGO**

```python
from sklearn.naive_bayes import BernoulliNB
classifier=BernoulliNB()
```

```python
from sklearn import metrics
import numpy as np
```

```python
classifier.fit(X_train, Y_train)
pred=classifier.predict(X_test)
score=metrics.accuracy_score(Y_test, pred)
print("Accuracy is: , ", score)
cm= metrics.confusion_matrix(Y_test, pred)
plot_confusion_matrix(cm, classes=["FAKE", "REAL"])
```

**PASSIVE AGGRESSIVE CLASSIFIER ALGO**

```python
from sklearn.linear_model import PassiveAggressiveClassifier
linear_clf=PassiveAggressiveClassifier(max_iter=50)
```

```python
linear_clf.fit(X_train, Y_train)
pred=linear_clf.predict(X_test)
pa_score=metrics.accuracy_score(Y_test, pred)
print("Accuracy is: , ", pa_score)
cm= metrics.confusion_matrix(Y_test, pred)
plot_confusion_matrix(cm, classes=["FAKE DATA", "REAL DATA"])
```

```python
input_user=input("Enter a news headline: ")
user_input_features = cv.transform([input_user])

prediction = linear_clf.predict(user_input_features)

if prediction[0] == 0:
    print("The news is fake.")
else:
    print("The news is real.")
```
